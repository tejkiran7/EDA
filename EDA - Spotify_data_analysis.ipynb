### Top 5 most popular artists

```python
top_five_artists = df.groupby("artist").count().sort_values(by="song_title",ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
```

### Top 5 loudest tracks

```python
top_5_loudest_tracks = df[["loudness","song_title"]].sort_values(by="loudness",ascending=True)[:5]
top_5_loudest_tracks
```

```python
plt.figure(figsize = (10, 5))
sns.barplot(y="song_title",x="loudness",data=top_5_loudest_tracks)
plt.title("Top 5 loudest Tracks")
plt.show()
```

### Artist with the most danceability song

```python
top_5_artists_danceable = df[["danceability","song_title","artist"]].sort_values(by="danceability",ascending=False)[:5]
top_5_artists_danceable
```

```python
import plotly.express as px
fig = px.bar(top_5_artists_danceable,x="artist",y="danceability")
fig.show()
```

### Top 10 Instrumental Tracks

```python
top_10_Instrumental_Tracks = df[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",ascending=False)[:10]
top_10_Instrumental_Tracks
```

```python
import plotly.express as px
fig = px.pie(top_10_Instrumental_Tracks, values='instrumentalness', names='song_title')
fig.show()
```

### Multiple feature plots

```python
interest_feature_cols = ["tempo","loudness",'acousticness', 'danceability', 'duration_ms', 
                         'energy','instrumentalness','liveness', 'speechiness', 'valence']
```

```python
for feature in interest_feature_cols:
    pos_data = df[df["target"]==1][feature]
    neg_data = df[df["target"]==0][feature]
    
    plt.figure(figsize=(12,7))
    
    sns.histplot(pos_data,bins=30,label='Positive',color='green',kde=True)
    sns.histplot(neg_data,bins=30,label='Negative',color='red',kde=True)
    
    plt.legend(loc="upper right")
    plt.title(f"Postive and Negative Histogram for {feature}")
    plt.show()
```
